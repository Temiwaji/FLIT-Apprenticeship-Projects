#### MARKET BASKET ANALYSIS USING APRIORI ALGORITHM

    5.0 FEATURE ENGINEERING

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# group items purchased per customer per day into a single basket (CustomerTransaction)

# Convert columns to string before concatenation
market_basketdata["CustomerTransaction"] = market_basketdata["Member_number"].astype(str) + "_" + market_basketdata["Date"].astype(str)
market_basketdata.head()
```

```python
# Convert Member_number to string
market_basketdata["Member_number"] = market_basketdata["Member_number"].astype(str)

# Check purchases of a unique customer
uniquecustomer_purchase = market_basketdata[market_basketdata["Member_number"] == "2552"].sort_values("Date")

uniquecustomer_purchase
```

```python
# Create a cross-tab of the CustomerTransaction and itemDescription columns from the DataFrame
basket = pd.crosstab(market_basketdata["CustomerTransaction"], market_basketdata["itemDescription"])
basket.head()
```

```python
# Convert each element to 1 if its value is greater than 0, otherwise to 0
apriori_basketdata = basket.applymap(lambda x: 1 if x > 0 else 0)
apriori_basketdata.head()
```

```python
# Apply the Apriori algorithm to generate frequent itemsets and association rules.
def generate_association_rules(data, min_support=0.005, metric="lift", top_n=20):
    frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric=metric)
    sorted_rules = rules.sort_values(by=metric, ascending=False).head(top_n)

    return sorted_rules
```

```python
apriori_basketdata = (basket > 0).astype(int)
top_rules = generate_association_rules(apriori_basketdata, min_support=0.005, metric="lift", top_n=20)
top_rules
```

```python
# Create a matrix of product associations
product_associations = top_rules.pivot(index="antecedents", columns="consequents", values="zhangs_metric").fillna(0)

# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(product_associations, annot=True, cmap="Blues", fmt=".1f", linewidths=.5)
plt.title("FREQUENTLY BOUGHT TOGETHER - ASSOCIATION RULES HEATMAP")

# Show the plot
plt.show()
```

**Observation**

A negative value indicates that the presence of the antecedent tends to decrease the probability of the consequent. Despite a certain level of confidence, the occurrence of the antecedent diminishes the overall support or likelihood of the consequent. This negative impact results in an anti-association between the antecedent and the consequent.

```python
# Taking pairs with positive Zhang's metric in a separate dataframe
positive_pairs = top_rules[top_rules["zhangs_metric"] > 0]

# Display the first few rows of the new DataFrame
positive_pairs.head()
```

```python
# Creating a matrix of product associations
positive_association = positive_pairs.pivot(index="antecedents", columns="consequents", values="zhangs_metric").fillna(0)

# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(positive_association, annot=True, cmap="Blues", fmt=".2f", linewidths=.5)
plt.title("Frequently Bought Together - Association Rules Heatmap")
plt.show()
```

Summary:

In examining the top 20 items, Whole milk emerges as the most frequently purchased product, leading the list. Following closely are items categorized under other vegetables and rolls/buns. A notable spike in purchases occurred in September 2015. However, monthly fluctuations indicate a decline in October for both 2014 and 2015, with the lowest recorded point in October 2015.

In the realm of association rule analysis, negative values in the rules signify a decrease in the likelihood of the consequent when the antecedent is present. This implies that despite a certain level of confidence, the occurrence of the antecedent diminishes the overall support or likelihood of the consequent. These negative impacts contribute to an anti-association between the antecedent and the consequent.

Recommendation:

Strategically, market efforts should concentrate on promoting Whole milk, given its consistent high demand. Additionally, a targeted approach to boost sales of other vegetables and rolls/buns could capitalize on their close popularity ranking.

Considering seasonal variations, leveraging the observed surge in September 2015 with targeted promotions can be a key strategy. Simultaneously, acknowledging the historical decline in October and implementing strategic adjustments will be crucial to mitigate potential drops in future years.

In the context of association rules, caution is advised when promoting items with negative association values.
